### 함수를 활용한 코드 통합
- 매개변수는 검색 키워드가 들어가도록 하고 return은 DF으로 반환 받도록 설정

```python
def blog_def(keyword: str) -> pd.DataFrame:
    title_list, name_list, date_list = [],[],[]

    url = f"https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}"
    U_A = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36"}
    res = req.get(url, headers = U_A)
    soup = bs(res.text, "lxml")

    title = soup.select("a > span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1.sds-comps-text-type-headline1.sds-comps-text-weight-sm")
    name = soup.select("div.sds-comps-horizontal-layout.sds-comps-inline-layout.sds-comps-profile-info-title > span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1")
    date = soup.select("div.sds-comps-horizontal-layout.sds-comps-inline-layout.sds-comps-profile-info-subtexts > span.sds-comps-text.sds-comps-text-type-body2.sds-comps-text-weight-sm.sds-comps-profile-info-subtext")

    for i in range(len(title)):
        title_list.append(title[i].text.strip())
        name_list.append(name[i].text.strip())
        date_list.append(date[i].text.strip())

            
    blog_df = pd.DataFrame({"타이틀": title_list,
                       "블로거명":name_list,
                       "게시일":date_list},
                      index = range(1,31))
    blog_df.index.name = "No"
        
    return blog_df
```

```python
blog_def("삼성전자")
```

### 검색창에 특정 단어 입력
- 컴퓨터에게 검색창을 찾게 명령하기 위해서는 다른 요소들과 마찬가지로 검색창의 태그 및 선택자를 알아야함

```python
# By: selenium에서 선택자를 구분해주는 함수
from selenium.webdriver.common.by import By
```

```python
# find_element: selenium에서 웹의 특정 요소하나를 찾아주는 함수(여러개를 찾으려면 s를 붙여 사용)
# CSS_SELECTOR: CSS선택지를 찾아주는 인자
search = driver.find_element(By.CSS_SELECTOR,"#query")
```

### XPATH 절대경로 사용하기
- XPATH는 원하는 요소에 정확히 접근하기 위한 절대경로를 가져오는 기능
- 하나의 요소에 정확하게 접근할 때 사용하면 편리한, 여러개의 요소에 접근해야 하거나 웹 페이지가 자주 바뀌는 유동적인 환경에서는 효율이 떨어짐
- 따라서 변동사항이 저긍ㄴ 개별 요소를 추출할 때 사용하면 매우 편리함(검색버튼, 메뉴버튼, 등 잘 바뀌지 않는 요소)
- 요소에 ID나 Class 값이 붙어있지 않아 구조나 텍스트 만으로 대상을 찾아야할 때 활용하면 좋음

```python
search = driver.find_element(By.ID,"query")
search.send_keys("오늘의 날씨")

btn = driver.find_element(By.XPATH,'//*[@id="search-btn-svg"]') # 절대결로로 접근
btn.click()
```

### 스크롤 기능 사용하기
- 스크롤을 위해서는 스크롤ㄹ이 가능한 전체 틀인 body 태그 자체를 선택한 뒤 스크롤 기능을 활용하느 것도 방법임

```python
# TAG_NAME 태그명으로 요소에 접근
body = driver.find_element(By.TAG_NAME, "body")

# END: 맨 아래쪽으로 이동(DOWN은 아래 방향 키보드 한번)
body.send_keys(Keys.END)
```

```python
# quit: 브라우저 창을 닫으며 연결이나 메모리도 모두 종료 시킴
driver.quit()

# close()도 닫는 기능이지만 메모리는 그대로 남아있음
```

```python
# 8) ver2
def Blog_title(keyword: str) -> pd.DataFrame:
    
    title_list = []
    url = f"https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}"
    U_A = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36"}
    res = req.get(url, headers = U_A)
    soup = bs(res.text, "lxml")

    title = soup.select("span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1.sds-comps-text-type-headline1.sds-comps-text-weight-sm")

    for i in range(len(title)):
        title_list.append(title[i].text.strip())

    Blog_df = pd.DataFrame({"블로그 제목": title_list},
                            index = range(1, len(title)+1))
    Blog_df.index.name = "No"

    return Blog_df
```
