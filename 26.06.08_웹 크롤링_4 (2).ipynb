### 페이지 하단의 '더보기' 버튼을 눌러 메뉴를 확장

```python
# 버튼 자체에 접근
btn = driver.find_element(By.CSS_SELECTOR, "a.c_05")

# 클릭
btn.click()
```

```python
# 더보기 버튼이 몇번이나 더 있을지 모르기 때문에 while문을 통해서 계속 눌러주자
while True :
    btn = driver.find_element(By.CSS_SELECTOR, "a.c_05")
    btn.click()

    # 2초간 대기했다가 다음 코드 실행
    time.sleep(2)

    # 더보기 버튼을 누르고 페이지가 뜨는데 시간이 약간 걸릴 수 있는데 대기 코드가 없다면 페이지가 뜨기도 전에
    # 무한정 버튼을 클릭하라고 명령하게 되어서 서버의 입장에서는 과부하가 걸릴 수 있고, 우리도 제대로 정보를 수집하지 못할 수 있음
    # 더 심한 경우에는 서버 자체에서 보안 공격으로 판단하고 접속한 클라이언트 PC의 IP주소를 차단할 수도 있음
```

### Python 예외 처리 문법(try-except문)
- 코드가 제대로 동작할때와 동작하지 않을때를 구분해주는 파이썬 문법
- 웹 페이지는 항상 같은 구조로 코드가 짜여지 있지는 않기 때문에 크롤링 시 예외처리를 많이 사용하게 됨
- **try문** : 일반적으로 동작하는 코드
- **except문** : try문이 동작하지 않는 예외의 경우(에러 발생시) 작동되는 코드

```python
# 일반적으로 동작하는 코드
try :
    while True :
        btn = driver.find_element(By.CSS_SELECTOR, "a.c_05")
        btn.click()
        time.sleep(2)

# try문이 동작하지 않는 예외의 경우에 동작하는 코드
except :
    print("더보기 버튼이 더이상 없네요 ㅠ_ㅠ 그래도 코드는 에러없이 이어서 동작합니다!")


# 더보기 버튼을 모두 누른 후(HTML요소를 모두 받아온 후) 메뉴와 가격정보를 추출
titles = driver.find_elements(By.CSS_SELECTOR, "h4.h.fz_03")
prices = driver.find_elements(By.CSS_SELECTOR, "div.item-price > strong")

titles_list, prices_list = [], []

for i in tqdm(range(len(titles))) :
    titles_list.append(titles[i].text)
    prices_list.append(prices[i].text + "원")

print("크롤링 완료!!!")
```

## 이제껏 작성했던 코드를 한 셀에 담아 DF생성까지 한번에 실행하기
- 웹 크롤러 실행파일을 개발한다면 코드가 한 셀에 시작부터 끝까지 모두 들어가 있어야 함

```python
# 크롬 드라이버 열어서 한솥 페이지 접근, 창 최대화
driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
driver.maximize_window()
time.sleep(2)

# try - except문으로 더보기 버튼 전부 누르기
try :
    while True :
        btn = driver.find_element(By.CSS_SELECTOR, "a.c_05")
        btn.click()
        time.sleep(2)
except :
    print("더보기 버튼이 더이상 없네요 ㅠ_ㅠ 그래도 코드는 에러없이 이어서 동작합니다!")

# 메뉴와 가격 정보에 접근
titles = driver.find_elements(By.CSS_SELECTOR, "h4.h.fz_03")
prices = driver.find_elements(By.CSS_SELECTOR, "div.item-price > strong")

# 메뉴와 가격 텍스트만 리스트에 넣기
titles_list, prices_list = [], []

for i in tqdm(range(len(titles))) :
    titles_list.append(titles[i].text)
    prices_list.append(prices[i].text + "원")

# DF로 만들기
hansot_dic = {"메뉴명": titles_list, "가격": prices_list}
hansot_df = pd.DataFrame(hansot_dic, index=range(1, len(titles_list)+1))
hansot_df.index.name = "No."
hansot_df
```

```python
# DF 전체 행 보기
pd.set_option("display.max_rows", None)   # None은 무한으로 출력하겠다는 뜻(출력할 행의 개수 입력 가능)
hansot_df
```

```python
# 구매가능 사이즈와 아닌 사이즈 추출
enabled_size, disabled_size = [], []

for i in size :
    input_id = i.get_attribute("id")

    # label 태그의 for 속성에 접근
    label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
    label_text = label.text.strip()   # 사이즈에 대한 텍스트만 추출

    # disabled가 없다면 None이므로 아래 조건문에서 False가 됨
    is_disabled = i.get_attribute("disabled")

    if is_disabled :
        disabled_size.append(label_text)
    else :
        enabled_size.append(label_text)

print("구매가능 사이즈 :", enabled_size)
print("구매불가 사이즈 :", disabled_size)
```

### 상위 4개 품목들의 정보 수집

```python
for i in tqdm(range(4)) :
    imgs = driver.find_elements(By.CSS_SELECTOR, "div.pro_thumbNail img.img_goods")
    imgs[i].click()
    time.sleep(3)

    title = driver.find_element(By.CSS_SELECTOR, "h2.title")
    mean_rating = driver.find_element(By.CSS_SELECTOR, "div.crema_product_reviews_score__container")
    size = driver.find_elements(By.CSS_SELECTOR, "#optSizeSection > ul.items > li > input")
    
    enabled_size, disabled_size = [], []
    
    for j in size :
        input_id = j.get_attribute("id")
    
        # label 태그의 for 속성에 접근
        label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
        label_text = label.text.strip()   # 사이즈에 대한 텍스트만 추출
    
        # disabled가 없다면 None이므로 아래 조건문에서 False가 됨
        is_disabled = j.get_attribute("disabled")
    
        if is_disabled :
            disabled_size.append(label_text)
        else :
            enabled_size.append(label_text)
    
    foot_width = driver.find_element(By.CSS_SELECTOR, "#optWidthSection > ul.items > li > label")
    cnt_reviews = driver.find_element(By.CSS_SELECTOR, "a.crema-product-reviews-count.crema-applied")

    print(title.text + " - " + mean_rating.text + " - " + foot_width.text + " - " + cnt_reviews.text[1:-1]+"개")
    print("구매가능 사이즈 :", enabled_size)

    driver.back()
    time.sleep(3)
```

### 한 셀에 모든 코드 넣어서 동작시켜보세요~!
- 상위 8개 품목에 대해서 순위를 매겨 DF로 만들어보세요

```python
# 크롬 열고 최대화 및 뉴발란스 사이트 이동
driver = wb.Chrome()
driver.maximize_window()
time.sleep(2)
driver.get('https://www.nbkorea.com/product/featuredList.action?cateGrpCode=250100&cIdx=1385')
time.sleep(2)
# 상품의 각 정보들을 담아줄 빈 리스트 생성(5개)
title_list, mean_rating_list, enabled_size_list, foot_width_list, cnt_reviews_list = [], [], [], [], []

# try-except문으로 8개의 상품의 세부정보에 접근
try :
    for i in tqdm(range(8)) :
        imgs = driver.find_elements(By.CSS_SELECTOR, "div.pro_thumbNail img.img_goods")
        imgs[i].click()
        time.sleep(3)
    
        title = driver.find_element(By.CSS_SELECTOR, "h2.title")
        title_list.append(title.text)
        
        mean_rating = driver.find_element(By.CSS_SELECTOR, "div.crema_product_reviews_score__container")
        mean_rating_list.append(mean_rating.text)
        
        size = driver.find_elements(By.CSS_SELECTOR, "#optSizeSection > ul.items > li > input")  
        enabled_size, disabled_size = [], []
        for j in size :
            input_id = j.get_attribute("id")
        
            # label 태그의 for 속성에 접근
            label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
            label_text = label.text.strip()   # 사이즈에 대한 텍스트만 추출
        
            # disabled가 없다면 None이므로 아래 조건문에서 False가 됨
            is_disabled = j.get_attribute("disabled")
        
            if is_disabled :
                disabled_size.append(label_text)
            else :
                enabled_size.append(label_text)
        enabled_size_list.append(enabled_size)
        
        foot_width = driver.find_element(By.CSS_SELECTOR, "#optWidthSection > ul.items > li > label")
        foot_width_list.append(foot_width.text)
        
        cnt_reviews = driver.find_element(By.CSS_SELECTOR, "a.crema-product-reviews-count.crema-applied")
        cnt_reviews_list.append(cnt_reviews.text[1:-1]+"개")

        driver.back()
        time.sleep(3)
        
except Exception as e :   # 에러 발생시 에러 출력
    print("에러명 :", e)

# 수집된 데이터 개수 확인(print문으로)
print("수집된 데이터 개수 확인 :")
print(len(title_list), len(mean_rating_list), len(cnt_reviews_list), len(enabled_size_list), len(cnt_reviews_list))

# DF 출력 코드
newbal_dic = {"상품명":title_list, "평점":mean_rating_list, "리뷰개수":cnt_reviews_list, 
              "구매가능 사이즈":enabled_size_list, "발볼":foot_width_list}
newbal_df = pd.DataFrame(newbal_dic, index=range(1, len(title_list)+1))
newbal_df.index.name = "순위"
newbal_df
```
