###  iframe태그
- 다른 서버의 웹 페이지 데이터나 리소스를 현재 서버읭 웹 페이지에 포함 시켜서 보여주는 HTML 태그
- 웹 페이지 내에 독립적인 작은 창으로 다른 서버의 데이터를 가져와 삽입하여 보여줄때 사용함
- 웹 페이지에 보여줄 데이터가 많고 실시간으로 변한다면 한번에 접급할떄 그만큼 소모되는 자원이 많아 속도가 느려질 수 있는데, 이를 방지하기 위해 데이터를 일부만 보여줄 수 있고 따로 관리가 가능한 iframe을 사용함(상품 판매 사이트에서 리뷰데이터와 같이 따로 관리해야하는 데이터의 양이 많은 경우 사용)

### 1~10page의 데이터를 모두 수집해보자
- 먼저 1page와 2,3page의 url을 비교해 규칙을 찾아보기(페이지 No와 관련된 정보가 있을 것)
- 일반적으로 개발자들이 여러 page가 있는 웹 사이트를 만들때 같은 기능은 최대한 유사한 코드를 활용함

```python
# tqdm : 반복문의 진행상황을 로딩바로 확인하게 해주는 함수
from tqdm import tqdm
```

```python
title_list, author_list, views_list = [],[],[]

# 1~10page까지의 데이터 수집을 위해 반복문 작성
for page in tqdm(range(1,11,1)):
    url = f"https://www.10000recipe.com/recipe/list.html?order=reco&page={page}"
    res = req.get(url, headers = U_A)
    soup = bs(res.text, "lxml")

    re_title = soup.select(" div.common_sp_caption_tit.line2")
    re_author = soup.select("div.common_sp_caption_rv_name > a")
    re_views = soup.select("div.common_sp_caption_rv > span.common_sp_caption_buyer")

    for a in range(40):
        title_list.append(re_title[a].text.strip())
        author_list.append(re_author[a].text.strip())
        views_list.append(re_views[a].text.strip()[4:])


re_df = pd.DataFrame({"레시피명":title_list,
                     "작성자":author_list,
                     "조회수":views_list},
                    index = range(1,len(title_list)+1))
re_df.index.name = "No"
re_df
```

### 특정 키워드가 들어간 레시피만 필터링

```python
# 레시피명 컬럼에서 '김치'가 포함된 행만 필터링
# str.contains: DF에서 특정 문자가 포함된 행 찾기
filtered_df = re_df[re_df["레시피명"].str.contains("감자")]

# reset_index: 인덱스 번호 재설정
# drop=True: 기존 인덱스 컬럼 삭제
filtered_df.reset_index(drop=True, inplace=True)
filtered_df.index = range(1, len(filtered_df)+1) # 필터링 된 DF의 행 개수만큼 인덱스 설정
filtered_df.index.name = "No"
filtered_df
```

### 함수로 코드 통합
- 매개변수는 검색할 최종 페이지, 필터링 할 검색어(디폴트는 검색어 없도록 설정)

```python
# 매개변수 뒤에 ': int' 와 같은 부분은 입력되는 매개변수의 자료형을 사전에 알려주는 것
# '-> pd.DataFrame' 의 경우 반환(return)되는 자료형을 사전에 알려주는 것
def search_recipe(end_page: int, search: str = "") -> pd.DataFrame :
    title_list, author_list, views_list = [], [], []

    for i in tqdm(range(1, end_page+1, 1)) :
        url = f"https://www.10000recipe.com/recipe/list.html?order=reco&page={i}"
        res = req.get(url, headers=U_A)
        soup = bs(res.text, "lxml")
    
        title = soup.select("div.common_sp_caption_tit.line2")
        author = soup.select("div.common_sp_caption_rv_name")
        views = soup.select("span.common_sp_caption_buyer")
    
        for i in range(len(title)) :
            title_list.append(title[i].text.strip())
            author_list.append(author[i].text.strip())
            views_list.append(views[i].text.strip()[4:])
    
    recipe_dic = {"레시피명": title_list, "작성자": author_list, "조회수": views_list}
    re_df = pd.DataFrame(recipe_dic, index=range(1,len(title_list)+1))

    # 검색어가 존재할 때(검색어를 입력했을 때)
    if search :
        # copy로 기존 DF를 복사해서 활용(원본 DF가 변형되는 것을 방지)
        filtered_df = re_df[re_df["레시피명"].str.contains(f'{search}')].copy()
    # 검색어가 따로 없을 때
    else :
        filtered_df = re_df.copy()

    filtered_df.reset_index(drop=True, inplace=True)
    filtered_df.index = range(1, len(filtered_df)+1)
    filtered_df.index.name= "No"
    return filtered_df
```

```python
search_recipe(5)
```
